### Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Data/Actual-Data/Actual_combine.csv")
df.head()
```

```python
# Checking the Null values if present in Dataset
sns.set(rc={'figure.figsize':(12,8)})
sns.heatmap(df.isnull(),yticklabels=False,cbar=True,cmap="summer")
print(df.isnull().value_counts())
```

```python
# Dropping the null values

df=df.dropna()
```

```python
# verifying elimination of null values
# df.isnull().value_counts()
```

```python
sns.pairplot(df)
```

```python
df.corr()
```

```python
# Visualizing the correlation using Heat Map

corrmat = df.corr()
top_corr_features = corrmat.index

corr_map=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap='twilight')
```

```python
df.corr().index
```

```python
# Independent features
X=df.iloc[:,:-1]
# Dependent features
y=df.iloc[:,-1]
```

### Feature Importance


```python
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.plot(kind='barh',colormap='summer')
```

#### Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

#### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
print("Coefficient of determination on R^2 for <--train set:{}".format(regressor.score(X_train,y_train)))
print("Coefficient of determination on R^2 for <--test set:{}".format(regressor.score(X_test,y_test)))
```

```python
from sklearn.model_selection import cross_val_score
score=cross_val_score(regressor,X,y,cv=10)
```
